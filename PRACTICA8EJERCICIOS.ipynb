```python
#EJERCICIO 1
import numpy as np
import matplotlib.pyplot as pl

x=[99,75,85,95,90,100,93,80,87,110]

y=[1485,900,1105,1330,1440,1200,1209,1200,1218,1650]
print(x)
n=len(x)
print("tama√±o= ",n)
print(y)

x=np.array(x)
y=np.array(y)
print(x)
print(y)
sumax=sum(x)
print("sumax= ",sumax)
sumay=sum(y)
print("sumay= ",sumay)
promx=sumax/n
print("promx= ",promx)
promy=sumay/n
print("promy= ",promy)
sumax2=sum(x*x)
print("sumax^2",sumax2)

sumaxy=sum(x*y)
print("sumaxy= ",sumaxy)
b=(sumax*sumay-n*sumaxy)/(sumax**2-n*sumax2)
a=promy-b*promx
print("b= ",b," a= ",a)
pl.plot(x,y,'o',label='Datos')
pl.plot(x,a+b*x,label='Ajuste')

pl.xlabel('x')
pl.ylabel('y')

pl.grid()
pl.legend()
pl.show()
```
